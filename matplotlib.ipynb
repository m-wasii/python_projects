```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_csv('US_Crime_Rates_1960_2014.csv')
```

```python
display(df)
```

```python
plt.scatter(x=df['Year'],y=df['Murder'])
plt.xlabel('Year')
plt.ylabel('Murder')
```

```python
plt.subplot(2,2,1)
x=df['Year']
y=df['Murder']
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
y=df['Larceny_Theft']
plt.plot(x,y,'g--')
plt.subplot(2,2,3)
y=df['Robbery']
plt.plot(x,y,'o')
plt.subplot(2,2,4)
y=df['Burglary']
plt.plot(x,y,'g-')
```

```python
x=df['Year']
y_murder = df['Murder']
y_burglary = df['Burglary']

plt.subplot(2, 1, 1)

plt.plot(x, y_murder,'r--')
plt.title('Murders')

plt.subplot(2, 1, 2)
plt.plot(x, y_burglary,'g--')
plt.title('Burglaries')

plt.show()
```
